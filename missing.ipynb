```python
import pandas as pd

# Load dataset
df = pd.read_csv("Employee.csv")
```

```python
df.isnull().sum()
```

```python
df = df.drop(columns=["Unnamed: 9", "Unnamed: 10", "Unnamed: 11"])
```

```python
df[df.isnull().any(axis=1)]
```

```python
df["Age"] = df["Age"].fillna(df["Age"].mean())
```

```python
df["ExperienceInCurrentDomain"] = df["ExperienceInCurrentDomain"].fillna(df["ExperienceInCurrentDomain"].median())
```

```python
df["EverBenched"] = df["EverBenched"].fillna(df["EverBenched"].mode()[0])
df["LeaveOrNot"] = df["LeaveOrNot"].fillna(df["LeaveOrNot"].mode()[0])
```
