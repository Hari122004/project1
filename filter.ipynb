```python
df["City"] = df["City"].str.strip()
df["Gender"] = df["Gender"].str.strip()
```

```python
df["City"] = df["City"].str.strip()
df["Gender"] = df["Gender"].str.strip()
df["grouph"] = df["PaymentTier"].astype(str).apply(lambda x: "Group H" if x == "3" else "Other")
df["ANALYSICS"] = pd.to_numeric(df["ExperienceInCurrentDomain"], errors="coerce").apply(
    lambda x: "Analytics" if x >= 3 else "Non-Analytics"
)

print(df[["grouph", "ANALYSICS"]].head())
```

```python
grouped = df.groupby(["JoiningYear", "City"]).size().reset_index(name="Total_Employees")
grouped
```

```python
left_emp = df[df["LeaveOrNot"] == 1] \
    .groupby(["JoiningYear", "City"]) \
    .size() \
    .reset_index(name="Left_Count")

left_emp
```

```python
avg_exp = df.groupby(["JoiningYear", "City"])["ExperienceInCurrentDomain"].mean().reset_index()
avg_exp
```
